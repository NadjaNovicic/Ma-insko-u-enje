```python
 import pandas as pd
 import seaborn as sns
 import matplotlib.pyplot as plt
 import numpy as np
 import math
 import operator
```

```python
 data = pd.read_csv("Real estate.csv")
 data.head()
```

```python
 Y=data['Y house price of unit area'].values
 X=data['X2 house age'].values
 plt.scatter(X,Y,color='red')
 plt.show()
```

```python
srednje_x = np.mean(X)
srednje_y = np.mean(Y)
```

```python

suma1 = 0
suma2 = 0
for i in range(414):
  suma1 += (X[i] - srednje_x) * (Y[i] - srednje_y)
  suma2 += (X[i] - srednje_x) ** 2
k = suma1 / suma2
n = srednje_y - (k * srednje_x)

print (f'k = {k} \nn = {n}')
```

```python
max_x = np.max(X)
min_x = np.min(X)

x = np.linspace(min_x, max_x)
y = k * x + n

plt.plot(x, y)
plt.scatter(X, Y,color='pink')

plt.xlabel('X2 house age')
plt.ylabel('Y house price of unit area')
plt.show()
```

Koristila sam nezavisnu promenljivu house age jer sam smatrala da to najvise utice na zavisnu promenljivu house price of unit area ,osvrnuci se na neke situacije iz stvarnog zivota.
